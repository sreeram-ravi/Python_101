The formula to determine the day of the week (for the Gregorian Caledar) is shown here:

* h is the day of the week (0 = Saturday, 1 = Sunday, 2 = Monday, …, 6 = Friday)
* q is the day of the month
* m is the month (3 = March, 4 = April, 5 = May, …, 14 = February)
* K is the year of the century (year % 100).
* J is the zero-based century (actually ⌊ year/100 ⌋)

```python
#@title Implementation

'''
This function takes in day, month, and year as the inputs.
And returns the day of the week (Saturday, Sunday, Monday... Friday)
'''

def find_day_of_week(day, month, year = 2022):
    # Defining numerical value with days
    Saturday = 0
    Sunday = 1
    Monday = 2
    Tuesday = 3
    Wednesday = 4
    Thrusday = 5
    Friday = 6

    #Modification done to the dates provided before calcuating 

    if ( month == 1 or month == 2 ):
      month = month + 12
      year = year - 1
    
    day_of_week = ( day +  13 * ( month + 1 )  // 5  + ( year % 100 ) + ( year % 100 ) // 4  + ( year // 100 ) // 4  +  5 * ( year // 100 )) % 7
    
    #day_of_week = int (day_of_week)

    # Modifications to the 
    
    if ( day_of_week == Saturday  ):
        return 'Saturday'
    elif ( day_of_week == Sunday ):
        return 'Sunday'
    elif ( day_of_week == Friday ):
        return 'Friday'
    elif ( day_of_week == Thrusday ):
        return 'Thrusday'
    elif ( day_of_week == Wednesday ):
        return 'Wednesday'
    elif ( day_of_week == Tuesday ):
        return 'Tuesday'
    elif ( day_of_week == Monday ):
        return 'Monday'    

# DO NOT CHANGE any of the following test code
# TEST DRIVEN PROGRAMMING

# Calling the function
x = find_day_of_week(6, 11, 2022)
print("\nTest Case 1: 6-Nov-2022: ", x)

if (x == "Sunday"):
  print("Test Case 1 passed")
else:
  print("Test Case 1 failed")

# Test Case 2
x = find_day_of_week(14, 10, 2022)
print("\nTest Case 2: 14-Oct-2022: ", x)

if (x == "Friday"):
  print("Test Case 2 passed")
else:
  print("Test Case 2 failed")


# Test Case 3
x = find_day_of_week(12, 11)
print("\nTest Case 3: 12-Nov-2022: ", x)

if (x == "Saturday"):
  print("Test Case 3 passed")
else:
  print("Test Case3 failed")


# Test Case 4
x = find_day_of_week(10, 11, 2021)
print("\nTest Case 4: 10-Nov-2021: ", x)

if (x == "Wednesday"):
  print("Test Case 4 passed")
else:
  print("Test Case 4 failed")


# Test Case 5
x = find_day_of_week(4, 1, 2021)
print("\nTest Case 5: 4-Jan-2021: ", x)

if (x == "Monday"):
  print("Test Case 5 passed")
else:
  print("Test Case 5 failed")


# Test Case 6
x = find_day_of_week(29, 2, 2020)
print("\nTest Case 6: 29-Feb-2020: ", x)

if (x == "Saturday"):
  print("Test Case 6 passed")
else:
  print("Test Case 6 failed")


# Test Case 7
x = find_day_of_week(25, 12, 2019)
print("\nTest Case 7: 25-Dec-2019: ", x)

if (x == "Wednesday"):
  print("Test Case 7 passed")
else:
  print("Test Case 7 failed")


# Test Case 8
# Add a Test for your birthday
x = find_day_of_week(1, 6, 1971)
print("\nTest Case 8: 01-Jun-1971: ", x)

if (x == "Tuesday"):
  print("Test Case 8 passed")
else:
  print("Test Case 8 failed")
```
